#### 1) React JS Developer

```python
# Read React JS Developer Resumes.

import os
import docx

# Change directory
path = r"D:\Project\Resumes\React JS Developer Resumes"

os.chdir(path)

# Read .docx file
react_js = []

def getText(file_path):
    doc = docx.Document(file_path)
    for para in doc.paragraphs:
        react_js.append(para.text)
              
# Iterate through all file to get file path.

for file in os.listdir():
    # Check whether file is in text format or not
    if file.endswith(".docx"):
        file_path = f"{path}\{file}" # Save file path
  
        #call read text file function
        getText(file_path) # Read file
        react_js.append("?") # seperate each resume in list using '?' in the end.
```

```python
#Stored sentances in list.
react_js
```

```python
#Join all sentances with space.

rtjs_list = []
rtjs_list.append(' '.join(react_js))
rtjs_list
```

```python
# Split resumes.

rtjs_rtext = rtjs_list[0].split("?")
```

```python
# Created new dataframe for React js resumes.

rtjs_resume_data = pd.DataFrame()
rtjs_resume_data['Resumes'] = rtjs_rtext
rtjs_resume_data['Category'] = 'React JS Developer'

# remove blank row.
rtjs_resume_data.drop(index=len(rtjs_resume_data['Resumes'])-1, axis=0, inplace=True)
rtjs_resume_data
```

####  Text Preprocessing

```python
#Preview of 1st resume data.
data['Resumes'][0]
```

```python
# Remove special characters, numbers and punctuations.

data['Cleaned_Resumes'] = data['Resumes'].str.replace("[^a-zA-Z]", " ")
data.head()
```

```python
data['Cleaned_Resumes'][0]
```

```python
# Remove links.

data['Cleaned_Resumes'] = data['Cleaned_Resumes'].str.replace('http[^\s][^s]+', " ")
data.head()
```

```python
data['Cleaned_Resumes'][0]
```

```python
#Text normalization.

data['Cleaned_Resumes'] = data['Cleaned_Resumes'].apply(lambda x: ' '.join(i.lower() for i in x.split()))
data.head()
```

```python
data['Cleaned_Resumes'][0]
```

```python
#Word frequency using bar chart.

freq = pd.Series(' '.join(data['Cleaned_Resumes']).split()).value_counts()[:20] # for top 20
ax = sns.barplot(x=freq.index, y=freq.values)
ax.bar_label(ax.containers[0]);
plt.xticks(rotation=90);
```

```python
#remove stopwords.

data['Cleaned_Resumes'] = data['Cleaned_Resumes'].apply(lambda x: ' '.join([i for i in x.split() if i not in stopwords.words('english')]))
data.head()
```

```python
data['Cleaned_Resumes'][0]
```

```python
# Lemmatization

lemma = WordNetLemmatizer()

data['Cleaned_Resumes'] = data['Cleaned_Resumes'].apply(lambda x: ' '.join(lemma.lemmatize(i) for i in x.split()))
data.head()
```

```python
data['Cleaned_Resumes'][0]
```

```python
freq = pd.Series(' '.join(data['Cleaned_Resumes']).split()).value_counts()
least_freq = freq[freq.values == 1].index
least_freq
```

```python
# Removing least frequent words.

data['Cleaned_Resumes'] = data['Cleaned_Resumes'].apply(lambda x: ' '.join(i for i in x.split() if i not in least_freq))
data.head()
```

```python
data['Cleaned_Resumes'][0]
```

```python
#Word frequency after removing stopwords and least frequent words.

freq = pd.Series(' '.join(data['Cleaned_Resumes']).split()).value_counts()[:20] # for top 20
ax = sns.barplot(x=freq.index, y=freq.values)
ax.bar_label(ax.containers[0]);
plt.xticks(rotation=90);
plt.tight_layout()
```

#### Feature Extraction

- There are 4 techniques used to convert textual data to numeric  
1) Binary Weights  
2) Bag-of-Words(BoW)  
3) TF-IDF Vectorizer  
4) Word Enbedding  

###### Feature extraction using TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()

x = tfidf.fit_transform(data['Cleaned_Resumes'])
x.toarray()
```

```python
tfidf.get_feature_names_out()
```

```python
df = pd.DataFrame(x.toarray(), columns=tfidf.get_feature_names_out())
df['Category'] = data['Category']
df
```

```python
# Label encoder for dependent variable.

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Category'] = le.fit_transform(df['Category'])
```

```python
# Data split into independent and dependent variable.

x = df.iloc[:,:-1]
y = df['Category']
```

#### Model Building

- Classification models

```python
#Libraries.
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

##### 1) Train Test Split

```python
# Train Test Split
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.2, random_state=1)
```

1) Logistic Regression

```python
lg_model = LogisticRegression()

lg_model.fit(xtrain,ytrain)
ypred = lg_model.predict(xtest)

lg_accuracy = lg_model.score(xtest,ytest)

print('Training Accuracy = ', lg_model.score(xtrain,ytrain))
print('Testing Accuracy = ', lg_accuracy)
```

2) KNN

```python
knn_model = KNeighborsClassifier()

knn_model.fit(xtrain,ytrain)
ypred = knn_model.predict(xtest)

knn_accuracy = knn_model.score(xtest,ytest)

print('Training Accuracy = ', knn_model.score(xtrain,ytrain))
print('Testing Accuracy = ', knn_accuracy)
```

3) Naive Bayes

```python
nb_model = MultinomialNB()

nb_model.fit(xtrain,ytrain)
ypred = nb_model.predict(xtest)

nb_accuracy = nb_model.score(xtest,ytest)

print('Training Accuracy = ', nb_model.score(xtrain,ytrain))
print('Testing Accuracy = ', nb_accuracy)
```

4) Decision Tree

```python
#Decision tree using C5.0

dt_model = DecisionTreeClassifier(criterion='entropy')

dt_model.fit(xtrain,ytrain)
ypred = dt_model.predict(xtest)

dt1_accuracy = dt_model.score(xtest,ytest)

print('Training Accuracy = ', dt_model.score(xtrain,ytrain))
print('Testing Accuracy = ', dt1_accuracy)
```

```python
#Decision tree using CART

dt_model = DecisionTreeClassifier(criterion='gini')

dt_model.fit(xtrain,ytrain)
ypred = dt_model.predict(xtest)

dt2_accuracy = dt_model.score(xtest,ytest)

print('Training Accuracy = ', dt_model.score(xtrain,ytrain))
print('Testing Accuracy = ', dt2_accuracy)
```

5) Random Forest

```python
#Random forest model using Decision tree C5.0

rf_model = RandomForestClassifier(criterion='entropy')

rf_model.fit(xtrain,ytrain)
ypred1 = rf_model.predict(xtest)

rf1_accuracy = rf_model.score(xtest,ytest)

print('Training Accuracy = ', rf_model.score(xtrain,ytrain))
print('Testing Accuracy = ', rf1_accuracy)
```

6) SVM

```python
svc = SVC()

svc.fit(xtrain,ytrain)
ypred = svc.predict(xtest)

svc_accuracy = svc.score(xtest,ytest)

print('Training Accuracy = ', svc.score(xtrain,ytrain))
print('Testing Accuracy = ', svc_accuracy)
```

- Accuracy scores obtained using train test split.

```python
Accuracy = pd.DataFrame({
                        'Model' : ['Logistic Regression','KNN','Naive Bayes','Decision tree(C5.0)','Decision tree(CART)','Random Forest(C5.0)','Random Forest(CART)','SVM'],
                        'Accuracy' : [lg_accuracy,knn_accuracy,nb_accuracy,dt1_accuracy,dt2_accuracy,rf1_accuracy,rf2_accuracy,svc_accuracy]})
Accuracy = Accuracy.sort_values(by='Accuracy', ascending=False)
Accuracy.reset_index(drop=True, inplace=True)
Accuracy
```

- KNN, Random Forest(C5.0) and Random Forest(CART) model shows 100% testing acuuracy, so these models have overfitting problem.
- Logistic Regression shows highest testing accuracy but this model is not good for multi-class classification.
- So using train test split we can go with Naive Bayes or SVM because both shows same testing accuracy.

##### 2) K-Fold Cross Validation

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=1)

lg_model = LogisticRegression()
lg_results = cross_val_score(lg_model, x, y, cv=kfold)

lg_accuracy = lg_results.mean()
print('Accuracy = ', lg_accuracy)
```
